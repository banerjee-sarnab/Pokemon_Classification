```python
import os
folders = os.listdir("D:\My Programs\Projects on CV and ML\Pokemon Classification\images")
print(folders)
```

```python
image_data = []
labels = []

label_dict = {
    "Abra" : 0,
    "Charmander" : 1,
    "Pikachu" : 2,
    "Squirtle" : 3
}
```

```python
import keras
# from keras.preprocessing import image

for ix in folders :
  path = os.path.join("D:\My Programs\Projects on CV and ML\Pokemon Classification\images", ix)
  for im in os.listdir(path) :
    img = keras.utils.load_img(os.path.join(path, im), target_size = (224,224))
    img_array = keras.utils.img_to_array(img)
    image_data.append(img_array)
    labels.append(label_dict[ix])
```

```python
print(len(image_data), len(labels))
```

```python
import random
import numpy as np
combined = list(zip(image_data, labels))
random.shuffle(combined)

image_data[:], labels[:] = zip(*combined)
```

```python
print(labels[:5])
```

```python
x_train = np.array(image_data)
y_train = np.array(labels)

print(x_train.shape, y_train.shape)
```

```python
from keras.utils import np_utils
y_train = np_utils.to_categorical(y_train)
print(x_train.shape, y_train.shape)
```

```python
# Create the Resnet-50 Model
from keras.applications.resnet import ResNet50
from keras.optimizers import Adam
from keras.layers import *
from keras.models import Model
from matplotlib import pyplot as plt
```

```python
model = ResNet50(include_top=False, weights='imagenet',input_shape=(224,224,3) )
```

```python
model.summary()
```

```python
# Adding New Classification layers to pre-trained model keeping the Conv Base intact
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256, activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(4, activation='softmax')(d1)

model_new = Model(inputs=model.input, outputs=fc2)
model_new.summary()
```

```python
adam = Adam(learning_rate=0.00003)
model_new.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
  print(ix, model_new.layers[ix])
```

```python
for ix in range(169) :
  model_new.layers[ix].trainable = False # Fine-tuning
model_new.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_new.summary()
```

```python
hist = model_new.fit(x_train, y_train,
                     shuffle = True,
                     batch_size=16,
                     epochs=5,
                     validation_split=0.20
                     )
```
